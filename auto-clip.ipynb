```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = './data/move_1s.csv'
data = pd.read_csv(file_path).to_numpy()

# Extract columns
col1 = data[:, 0]
col2 = data[:, 1]
col3 = data[:, 2]
```

```python
plt.plot(range(len(data)), data[:, 0])
plt.plot(range(len(data)), data[:, 1])
plt.plot(range(len(data)), data[:, 2])
plt.show()
```

```python
def auto_clip(data):
    compared_data = data[0, :3]
    
    left_idx = 0
    for i in range(len(data)):
        distance = np.linalg.norm(data[i, :3] - compared_data)
        left_idx = i
        if (distance > 0.5):
            break
    left_idx = max(0, left_idx - 3)
    return data[left_idx:, :]

clipped_data = auto_clip(data)
plt.plot(range(len(clipped_data)), clipped_data[:, 0])
plt.plot(range(len(clipped_data)), clipped_data[:, 1])
plt.plot(range(len(clipped_data)), clipped_data[:, 2])
plt.show()
```
